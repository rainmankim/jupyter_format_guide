# <span style='background :white' >   <span style='color:darkblue'> <u> Essential Keyboard Shortcuts </span></span> <a id='key_shortcuts'>

<div class="alert alert-block alert-info">
<b><u>Two Modes in Jupyter</u></b> 
    
- EDIT mode: when cursor is blinking inside codeblock
- COMMAND mode: when cursor is NOT blinking inside codeblock

## <span style='background :navy' >    <span style='color:white'> Edit Mode Keyboard Shortcuts   </span>     </span> <a id='edit_mode_shortcuts'>

Edit Mode Example



- ```Ctrl + Enter``` : Run current cell
- ```Shift + Enter``` : Run current cell and move to next cell
- ```Alt + Enter``` : Run current cell and insert a cell
- ```Esc``` : Transition from EDIT Mode to COMMAND Mode

## <span style='background :navy' >    <span style='color:white'> Command Mode Shortcuts   </span>     </span> <a id='command_mode_shortcuts'>

- ```Ctrl + Enter``` : Run current cell
- ```Shift + Enter``` : Run current cell and move to next cell
- ```Alt + Enter``` : Run current cell and insert a cell
- ```Enter``` :  Transition from COMMAND Mode to EDIT Mode 
- ```up down key```: Navigate up and down
- ```space``` : Scroll down
- ```shit + space``` : Scroll up
- ```b```: Insert a new cell below current cell
- ```a```:  Insert a new cell above current cell
- ```dd```: Delete current cell
- ```z``` : Undo cell deletion
- ```c``` : Copy current cell
- ```x``` : Cut current cell
- ```v``` : Paste current cell
- ```m``` : Change cell to Markdown
- ```y``` : Change cell to Code
- ```p``` : Open up Command Palette
- ```h``` : See full list of keyboard shortcuts
